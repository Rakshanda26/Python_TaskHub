```python
import os
import re
import docx
import PyPDF2
from striprtf.striprtf import rtf_to_text
import pandas as pd

# Define the directory path
directory_path = 'D:\\Machine_Learning_Final_All\\Python_Task_All\\pandas-main\\same-resume-year-wise-master'

class PatternSearcher:
    
    def __init__(self, src):
        self.src = src
        
    def find_email(self):
        pattern = re.compile('[a-z0-9]+[\._]?[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}')
        match = re.search(pattern, self.src)
        if match:
            return match.group()
        else:
            return "No email found"
    
    def find_linkedin(self):
        pattern = re.compile('www.linkedin.com/in/[a-zA-Z-]+[a-z0-9]+')
        match = re.search(pattern, self.src)
        if match:
            return match.group()
        else:
            return "No LinkedIn found"
    
    def find_git(self):
        pattern = re.compile('github.com/[a-zA-Z0-9]+')
        match = re.search(pattern, self.src)
        if match:
            return match.group()
        else:
            return "No GitHub found"

# Define a list of skills
skills = ["Python", "Java", "Machine Learning", "Data Analytics", "SQL", "Excel", "Team Leadership", "Tableau", "PowerBI", "MongoDB"]

def extract_skills(resume_text):
    extracted_skills = []

    for skill in skills:
        pattern = re.compile(r'\b' + re.escape(skill) + r'\b', re.IGNORECASE)
        if re.search(pattern, resume_text):
            extracted_skills.append(skill)

    return extracted_skills

df = pd.DataFrame(columns=["resume_name", "email id", "linkedIn id", "git id", "extracted_skills"])

for file in os.listdir(directory_path):
    if file.endswith(".docx") or file.endswith(".doc"):
        try:
            docObj = docx.Document(os.path.join(directory_path, file))
            doc_str = '\n'.join([para.text for para in docObj.paragraphs])
        except Exception as e:
            print(f"Skipped {file} - {e}")
            continue
    elif file.endswith(".pdf"):
        f = open(os.path.join(directory_path, file), 'rb')
        pdfread = PyPDF2.PdfReader(f)
        doc_str = ''
        for page in pdfread.pages:
            doc_str += page.extract_text()
    elif file.endswith(".rtf"):
        f = open(os.path.join(directory_path, file))
        content = f.read()
        doc_str = rtf_to_text(content)
    else:
        print("Skipped:", file)
        continue

    buff_obj = PatternSearcher(doc_str)
    email_id = buff_obj.find_email()
    linkedin_id = buff_obj.find_linkedin()
    git_id = buff_obj.find_git()

    # Extract skills from the resume
    extracted_skills = extract_skills(doc_str)

    new_row = {"resume_name": file, "email id": email_id, "linkedIn id": linkedin_id, "git id": git_id, "extracted_skills": extracted_skills}
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# Save the results to a CSV file
df.to_csv('resume_info_with_skills11.csv', index=False)
```
