### 1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t

```python
def find_the_difference (s,t):
    """This is a function to identify the difference letter between two strings"""
    
    dict_s={}
    dict_t={}
    
    if len(s)==0:
        return (t)

    else:
        for i in t:          # Creating a dictonary for string s containing keys from string t assigned to 0
            dict_s[i]=0
    
        for i in s:          # Increasing the value associated with keys for dictonary s based on character occurances in s
            dict_s[i]=dict_s[i]+1
    
        for i in t:         # Creating a dictonary for string t containing keys from string t assigned to 0
            dict_t[i]=0
    
        for i in t:         # Increasing the value associated with keys for dictonary t based on character occurances in t
            dict_t[i]=dict_t[i]+1
    
        for i in dict_t:    # Comparing both the dictonaries for values of keys and detecting mismatch as additional letter
            if dict_t[i]!= dict_s[i]:
                return(i)
    
```

### 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

List order is:

[int, str, bool, list, tuple, dictionary]


```python
def count_datatypes (*args):
    """This is a function to count occurances of distinct data types in a given list"""
    dict_types={int:0, str:0, bool:0, list:0, tuple:0, dict:0}
    list_out=[]
    
    for i in args:
        for j in dict_types:
            if type(i) == j:
                dict_types[j]= dict_types[j]+1
        
    for i in dict_types:
        list_out.append(dict_types[i])
    
    return list_out
        
```

```python
count_datatypes(1, 45, "Hi", False) 
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) 
```

### 3. Create a function to return desired number of terms of Fibonacci string generated based on given characters

```python
def fib_str (n,l):
    """This is a function to generate first n terms of Fibonacci string generated from given characters"""
    s1=l[0]
    s2=l[1]
    fibo=[]

    fibo.append(s1)
    fibo.append(s2)

    for i in range(2,n):
        fibo.append(fibo[i-1]+fibo[i-2])
    
    return fibo
```

```python
fib_str(3, ["j", "h"]) 
```

```python
fib_str(5, ["e", "a"])
```

```python
fib_str(6, ["n", "k"]) 
```

### 4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

"nines:your answer, threes:your answer, ones:your answer"

You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends)


```python
def ones_threes_nines (no):
    """This is a function to give the split of entered number in number of once, threes & nines"""
    if no<27 and no>=0:
        onces=0
        threes=0
        nines=0

        while (1*onces+3*threes+9*nines != no):
            if no>= 1*onces+3*threes+9*nines+9:
                nines=nines+1
            elif no>= 1*onces+3*threes+9*nines+3:
                threes=threes+1
            else:
                onces=onces+1

        out_dict={'nines':nines,'threes':threes,'ones':onces}
        return(out_dict)
    else:
        return "Enter a number between 0 to 26"
```

```python
ones_threes_nines(10) 
```

```python
ones_threes_nines(15)
```

```python
ones_threes_nines(22) 
```

```python
ones_threes_nines(28) 
```

### Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

```python
def fib(n):
    """This is a non recursive function to give required indexed term of fibonacci series"""
    fibo=[]
    fibo.append(0)
    fibo.append(1)
    for i in range (2,n+1):
        fibo.append(fibo[i-1]+fibo[i-2])
    return fibo[n]
```

```python
fib(6) 
```

```python
fib(1) 
```

```python
fib(2) 
```
