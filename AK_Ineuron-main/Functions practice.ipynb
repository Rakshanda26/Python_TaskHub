### Q1. Function for accepting a string and returning its length

```python
def f1 (s):
    """This is a function to compute length of the string without built in function """
    count = 0
    for i in s:
        count = count+1
    return count
```

```python
s1= input("Enter the string whos length to be measured:")
length= f1(s1)
print("Length of the entered string is:",length)
```

### Q2. Function to print index of every element of the object that is passed

```python
def f2 (obj):
    """This is a function to display indices of every element of a passed string/list/tuple/dictonary"""
    if type(obj)==list or type(obj)==str or type(obj)==tuple:
        for i in range(len(obj)):
            print ("Index of the element with value",obj[i],"is",i)
    elif type(obj)==dict:
        key=list(obj.keys())
        for j in range(len(obj)):
            print("Index of the key element",key[j],"is",j)
```

### Q3. Function taking 2 level nested dictonary as input and returning all the values in a list 

```python
def f3 (d):
    """This is a function to accept a 2 level nested dictonary and return list of all values"""
    val_list=[]
    for i  in d:
        if str(type(d[i]))!="<class 'dict'>":
            val_list.append(d[i])
        else:
            d1=d[i]
            for j in d1:
                if str(type(d1[j]))!="<class 'dict'>":
                    val_list.append(d1[j])
                else:
                    d2=d1[j]
                    for k in d2:
                        val_list.append(d2[k])
    return val_list
```

### Q11. A function to read and play video

```python
import cv2
```

```python
def vid_display (obj):
    """This is a function to read and display a video"""
    vid=cv2.VideoCapture(obj)
    while(vid.isOpened()):
        ret, frame = vid.read()
        if ret == True:
            cv2.imshow('Frame',frame)
        else:
            break
    vid.release()
```

### Q14. A function to access email

```python
import imaplib, email, base64
```

```python
def f14 (user,password,imap_url):
    """This is a function to access emails"""
    con = imaplib.IMAP4_SSL(imap_url)
    con.login(user, password)
    con.select('Inbox')
    
    data = con.search(None, 'ALL')
    mail_ids = data[1]
    id_list = mail_ids[0].split()
    
    first_email_id = int(id_list[0])
    latest_email_id = int(id_list[-1])
    
    for i in range(latest_email_id,first_email_id, -1):
            data = con.fetch(str(i), '(RFC822)' )
            for response_part in data:
                arr = response_part[0]
                if isinstance(arr, tuple):
                    msg = email.message_from_string(str(arr[1],'latin1'))
                    email_subject = msg['subject']
                    email_from = msg['from']
                    print('From : ' + email_from + '\n')
                    print('Subject : ' + email_subject + '\n\n')
```

### Q15. A function to send mails to desired receivers

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
```

```python
def f15 (sender_address,sender_pass,receiver_address,mail_content):
    """This is a function to send mail to an intended receiver"""
    message = MIMEMultipart()
    message['From'] = sender_address
    message['To'] = receiver_address
    message['Subject'] = 'A test mail sent by Python'
    
    message.attach(MIMEText(mail_content, 'plain'))
    session = smtplib.SMTP('smtp.gmail.com', 587)
    session.starttls()
    session.login(sender_address, sender_pass) 
    text = message.as_string()
    session.sendmail(sender_address, receiver_address, text)
    session.quit()
    print('Mail Sent')
```

```python
mail_content = '''Hello,
This is a mail sent using Python SMTP library.
Thank You'''
sender_address = 'abhikulkarni20@gmail.com'
sender_pass = 'nkppmgcdletdcpkx'
receiver_address = 'abhaykulkarni567@gmail.com'
f15(sender_address ,sender_pass ,receiver_address,mail_content)
```

### Q16. A function to read PDF files

```python
import PyPDF2
```

```python
def f16 (Add):
    """This is a function to read a PDF file"""
    pdfFileObj = open(Add, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    for i in range(pdfReader.numPages):
        pageObj = pdfReader.getPage(i)
        print(pageObj.extractText())
    pdfFileObj.close()
    
```

### Q20. A function to append two PDF files

```python
import PyPDF2
```

```python
def f20 (file1,file2):
    """This is a function to merge two PDF files given as input"""
    pdf1File = open(file1, 'rb')
    pdf2File = open(file2, 'rb')
    pdf1Reader = PyPDF2.PdfFileReader(pdf1File)
    pdf2Reader = PyPDF2.PdfFileReader(pdf2File)
    pdfWriter = PyPDF2.PdfFileWriter()
    for pageNum in range(pdf1Reader.numPages):
        pageObj = pdf1Reader.getPage(pageNum)
        pdfWriter.addPage(pageObj)
    for pageNum in range(pdf2Reader.numPages):
        pageObj = pdf2Reader.getPage(pageNum)
        pdfWriter.addPage(pageObj)
    pdfOutputFile = open('E:/IITM DoMS MS Systems IT/Sem1/MergedFiles.pdf', 'wb')
    pdfWriter.write(pdfOutputFile)
    pdfOutputFile.close()
    pdf1File.close()
    pdf2File.close()
```
