```python
import os
import collections
import re

directory_path = 'D:\\Machine_Learning_Final_All\\Python_Task_All'

def count_words_in_file(file_path):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()
        words = re.findall(r'\b[a-zA-Z]+\b', text)
        return collections.Counter(words)

word_counts = collections.Counter()
file_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path) if filename.endswith('.txt')]

for file_path in file_paths:
    word_count = count_words_in_file(file_path)
    word_counts.update(word_count)

word_counts_list = list(zip(word_counts.keys(), word_counts.values()))

print(word_counts_list)
```

```python
import os
import re

directory_path = 'D:\\Machine_Learning_Final_All\\Python_Task_All'

word_counts = {}
file_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path) if filename.endswith('.txt')]

for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()
        words = re.findall(r'\b[a-zA-Z]+\b', text)
        for word in words:
            initial = word[0]
            word_counts[initial] = word_counts.get(initial, 0) + 1

word_counts_list = list(word_counts.items())

print(word_counts_list)
```

```python
import os
import re

directory_path = 'D:\\Machine_Learning_Final_All\\Python_Task_All'

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        with open(os.path.join(directory_path, filename), 'r', encoding='utf-8', errors='ignore') as file:
            text = file.read()

            # Extract and print words without leading numbers
            words = re.findall(r'\b[^0-9\s][\w-]+\b', text)
            
            for word in words:
                print(word)
```
